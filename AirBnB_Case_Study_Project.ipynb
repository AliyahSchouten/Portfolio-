```python
import pandas as pd
import numpy as np
data=pd.read_csv(r"C:\Users\ALIYA\OneDrive\Documents\Coding_Temple\Week 3\Project\AB_NYC_2019.csv")
```

```python
display(data.sort_values("calculated_host_listings_count", ascending=False).drop_duplicates("host_name").head(10))

#The bisiest places are all in Manhattan and seem to be just host doing rents for apts

#Which jpsts are the busiest and why
ab_avg=data.groupby(['host_id','host_name'],as_index=False).sum(True).sort_values('reviews_per_month',ascending=False)

#% avaliablity not vacant would be a best metric 

ab_avg[['host_id','host_name','reviews_per_month']].head(5)

#Calculated host listings is total listings made by specific hosts, we did not feel
```

```python
# How many neighborhood groups are available and which shows up the most?

data_neighborhood=data.groupby(['neighbourhood_group','host_id'],as_index=False)
data_neighborhood
data.describe(exclude='number')
data.groupby('neighbourhood_group',as_index=False).count().sort_values('id', ascending=False)[['neighbourhood_group','id']]
#data.groupby('neighbourhood_group',as_index=False).sum().sort_values('reviews_per_month', ascending=False)[['neighbourhood_group','calculated_host_listings_count']]

```

```python
# Are private rooms the most popular in manhattan?
data = data[["neighbourhood_group", "room_type",'reviews_per_month']].groupby(["neighbourhood_group", 'room_type'],as_index=False).sum(True).sort_values('reviews_per_month',ascending =False)

data[data["neighbourhood_group"] == "Manhattan"]
#Entire home/apt is the most populat
```

```python
# Which hosts are the busiest and based on their reviews?
host_reve_avg=data.groupby(['host_id','host_name'],as_index=False).sum(True).sort_values('number_of_reviews',ascending=False)
host_reve_avg[['host_id','host_name','number_of_reviews']].head(5)



```

```python
#Which neighorhood group has the highest average price?

data.groupby('neighbourhood_group',as_index=False).mean(True).sort_values('price', ascending=False)[['neighbourhood_group','price']].round(2)
```

```python
# Which neighbor hood group has the highest total price?
data.groupby('neighbourhood_group',as_index=False).sum(True).sort_values('price', ascending=False)[['neighbourhood_group','price']].round(2)
```

```python
#Which top 5 hosts have the highest total price?
data.groupby(['host_id',"host_name"],as_index=False).sum(True).sort_values('price',ascending=False)[['host_id', 'host_name','price']].head()
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
zero_aval=data.groupby(['id','number_of_reviews','availability_365'],as_index=False).count().sort_values('availability_365')[['id','number_of_reviews',"availability_365"]]
mask2=(zero_aval["number_of_reviews"]>=100)&(zero_aval['availability_365']==0)
zero_aval[mask2]




```

```python
# What host has the highest total of prices and where are they located?
data.groupby(['host_id',"host_name",'neighbourhood_group'],as_index=False).sum(True).sort_values('price',ascending=False)[['host_id', 'host_name','price','neighbourhood_group']].head()

```

```python
# When did Danielle from Queens last receive a review?

dan_data = data[(data['host_name']== "Danielle")&(data['neighbourhood_group']=='Queens')&(data['last_review'].isnull() == False)].sort_values("last_review",ascending=False)
dan_data= dan_data.drop_duplicates("host_id")
dan_data[['host_id','host_name','neighbourhood_group','last_review']]
```

## Further Questions

1. Which host has the most listings?

```python
data.groupby(['host_id',"host_name"],as_index=False).sum(True).sort_values('calculated_host_listings_count',ascending=False)[['host_id', 'host_name','calculated_host_listings_count']].head()
```

2. How many listings have completely open availability?

```python
zero_aval=data.groupby(['id','calculated_host_listings_count','availability_365'],as_index=False).count().sort_values('availability_365')[['id','calculated_host_listings_count',"availability_365"]]
mask2=(zero_aval['availability_365']==365)
zero_aval[mask2].count()
```

3. What room_types have the highest review numbers?

```python
data[["neighbourhood_group", "room_type",'reviews_per_month']].groupby([ 'room_type'],as_index=False).sum(True).sort_values('reviews_per_month',ascending =False)
```
